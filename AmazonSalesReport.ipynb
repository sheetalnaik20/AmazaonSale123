```python
#Step1
#Convert the 'Date' column to a datetime object for time-series analysis.
df['Date'] = pd.to_datetime(df['Date'])
df.head(5)
```

```python
#Step2
#Calculate key metrics like total sales, total quantity, total orders, and average order value.
#Total Sales and Quantity:

#Calculate the total sales amount.
#Calculate the total quantity sold.
#Average Order Value (AOV):

#Calculate the Average Order Value by dividing the total sales amount by the total number of orders.
# Total Sales and Quantity
total_sales = df['Amount'].sum()
total_quantity = df['Qty'].sum()

# Total Orders
total_orders = df['Order ID'].nunique()

# Average Order Value (AOV)
average_order_value = total_sales / total_orders

print(f"Total Sales: ${total_sales}")
print(f"Total Quantity Sold: {total_quantity}")
print(f"Total Orders: {total_orders}")
print(f"Average Order Value: ${average_order_value:.2f}")
```

```python
#Step3
#Sales by Fulfilment Method and Sales Channel
#Objective: Summarize sales amounts and quantities for each fulfilment method and sales channel.
#Sales by Fulfilment Method:
#Group by 'Fulfilment' and calculate total sales and quantities.
#Sales by Sales Channel:
#Group by 'Sales Channel' and calculate total sales and quantities.
# Sales by Fulfilment Method
sales_by_fulfilment = df.groupby('Fulfilment')['Amount'].sum()
quantity_by_fulfilment = df.groupby('Fulfilment')['Qty'].sum()

print("Sales by Fulfilment Method:")
print(sales_by_fulfilment)
print("Quantity by Fulfilment Method:")
print(quantity_by_fulfilment)

# Sales by Sales Channel
sales_by_channel = df.groupby('Sales Channel')['Amount'].sum()
quantity_by_channel = df.groupby('Sales Channel')['Qty'].sum()

print("Sales by Sales Channel:")
print(sales_by_channel)
print("Quantity by Sales Channel:")
print(quantity_by_channel)
```

```python
#Step4
#Sales by Category
#Objective: Summarize sales amounts and quantities for each product category.
#Sales by Category:
#Group by 'Category' and calculate total sales and quantities.
# Sales by Category
sales_by_category = df.groupby('Category')['Amount'].sum()
quantity_by_category = df.groupby('Category')['Qty'].sum()

print("Sales by Category:")
print(sales_by_category)
print("Quantity by Category:")
print(quantity_by_category)
```

```python
#Step 5: Identify Trends and Patterns Over Time
#Objective: Aggregate sales data by month and year to identify trends.
#Create Year-Month Column:
#Extract year and month from 'Date' and create a 'YearMonth' column.
#Monthly Sales Trend:
#Group by 'YearMonth' and calculate total sales and quantities.
# Create YearMonth column
df['YearMonth'] = df['Date'].dt.to_period('M')

# Monthly Sales Trend
monthly_sales = df.groupby('YearMonth')['Amount'].sum()
monthly_quantity = df.groupby('YearMonth')['Qty'].sum()

print("Monthly Sales:")
print(monthly_sales)
print("Monthly Quantity Sold:")
print(monthly_quantity)
```

```python
#Step 6: Visualize Sales Trends
#Objective: Create visualizations to show sales trends over time.
#Total Sales Over Time:
#Line chart showing total sales amount over time (monthly).
#Total Quantity Sold Over Time:
#Line chart showing total quantity sold over time (monthly).
import matplotlib.pyplot as plt

# Total Sales Over Time
plt.figure(figsize=(10, 6))
monthly_sales.plot()
plt.title('Total Sales Over Time')
plt.xlabel('Date')
plt.ylabel('Sales Amount')
plt.show()

# Total Quantity Sold Over Time
plt.figure(figsize=(10, 6))
monthly_quantity.plot()
plt.title('Total Quantity Sold Over Time')
plt.xlabel('Date')
plt.ylabel('Quantity Sold')
plt.show()
```

```python
#Step 7: Summarize Fulfillment Methods
#Objective: Summarize the sales amount and quantities for each fulfillment method to understand their contribution to overall sales.
#Total Sales and Quantity by Fulfillment Method:
#Group by 'Fulfillment' and calculate total sales and quantities.
# Sales by Fulfillment Method
sales_by_fulfillment = df.groupby('Fulfilment')['Amount'].sum()
quantity_by_fulfillment = df.groupby('Fulfilment')['Qty'].sum()

print("Sales by Fulfillment Method:")
print(sales_by_fulfillment)
print("Quantity by Fulfillment Method:")
print(quantity_by_fulfillment)
```

```python
#Step 8: Analyze Effectiveness of Fulfillment Methods
#Objective: Investigate the effectiveness of fulfillment methods in delivering orders. This might include looking at delivery status and times if available.
#Average Sales per Order by Fulfillment Method:
#Calculate the average sales per order for each fulfillment method.
#Fulfillment Method Performance:

#If data on delivery status or times is available, analyze the success rate and average delivery time for each fulfillment method.
# Average Sales per Order by Fulfillment Method
average_sales_per_order = df.groupby('Fulfilment')['Amount'].mean()

print("Average Sales per Order by Fulfillment Method:")
print(average_sales_per_order)
```

```python
#Step 9: Delivery Status Analysis (if available)
#Objective: If 'Courier Status' or similar delivery status data is available, analyze the success rates of each fulfillment method.
#Successful Deliveries by Fulfillment Method:
#Calculate the percentage of successful deliveries for each fulfillment method.
if 'Courier Status' in df.columns:
    # Assuming 'Courier Status' has values like 'Delivered', 'Pending', etc.
    successful_deliveries = df[df['Courier Status'] == 'Delivered'].groupby('Fulfilment')['Order ID'].count()
    total_deliveries = df.groupby('Fulfilment')['Order ID'].count()
    success_rate = (successful_deliveries / total_deliveries) * 100

    print("Success Rate of Deliveries by Fulfillment Method:")
    print(success_rate)
```

```python
#Step 10: Visualize Fulfillment Method Analysis
#Objective: Create visualizations to compare the performance of different fulfillment methods.
#Sales by Fulfillment Method:
#Bar chart showing total sales amount for each fulfillment method.
#Quantity by Fulfillment Method:
#Bar chart showing total quantity sold for each fulfillment method.
#Success Rate of Deliveries by Fulfillment Method (if available):
#Bar chart showing the success rate of deliveries for each fulfillment method.
import matplotlib.pyplot as plt

# Sales by Fulfillment Method
plt.figure(figsize=(10, 6))
sales_by_fulfillment.plot(kind='bar')
plt.title('Sales by Fulfillment Method')
plt.xlabel('Fulfillment Method')
plt.ylabel('Sales Amount')
plt.show()

# Quantity by Fulfillment Method
plt.figure(figsize=(10, 6))
quantity_by_fulfillment.plot(kind='bar')
plt.title('Quantity by Fulfillment Method')
plt.xlabel('Fulfillment Method')
plt.ylabel('Quantity Sold')
plt.show()


```

```python
#Step11;Calculate Customer Metrics
#Objective: Calculate metrics like total amount spent, total quantity bought, and number of orders for each customer.
#Aggregate Customer Data:
#Group data by 'Order ID' and calculate total amount spent, total quantity bought, and number of orders.
# Aggregate Customer Data
customer_data = df.groupby('Order ID').agg({
    'Amount': 'sum',
    'Qty': 'sum',
    'Order ID': 'count',
    'ship-city': 'first',
    'ship-state': 'first',
    'ship-country': 'first'
}).rename(columns={'Order ID': 'Order Count'})

customer_data.reset_index(inplace=True)
customer_data.head()
```

```python
# Location Segmentation
location_segments = customer_data.groupby(['ship-city', 'ship-state', 'ship-country']).agg({
    'Amount': 'sum',
    'Qty': 'sum',
    'Order Count': 'sum'
}).sort_values(by='Amount', ascending=False)

location_segments.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Spender Segments
plt.figure(figsize=(10, 6))
sns.countplot(data=customer_data, x='Spender Segment')
plt.title('Customer Distribution by Spender Segment')
plt.xlabel('Spender Segment')
plt.ylabel('Number of Customers')
plt.show()


```

```python
 #Aggregate Sales Data by Location
#Objective: Calculate total sales and quantities for each city and state.

#Group by State and City:
#Group the data by 'ship-state' and 'ship-city' to calculate total sales and quantities.
# Aggregate Sales Data by State
sales_by_state = df.groupby('ship-state').agg({
    'Amount': 'sum',
    'Qty': 'sum'
}).sort_values(by='Amount', ascending=False)

# Aggregate Sales Data by City
sales_by_city = df.groupby('ship-city').agg({
    'Amount': 'sum',
    'Qty': 'sum'
}).sort_values(by='Amount', ascending=False)

print("Sales by State:")
print(sales_by_state.head())

print("Sales by City:")
print(sales_by_city.head())
```

```python
#Visualize Geographical Distribution
#Objective: Create visualizations to show the distribution of sales by city and state.
#Sales by State:
#Bar chart showing total sales amount for each state.
#Sales by City:
#Bar chart showing total sales amount for each city.
import matplotlib.pyplot as plt

# Sales by State
plt.figure(figsize=(10, 6))
sales_by_state['Amount'].plot(kind='bar')
plt.title('Sales by State')
plt.xlabel('State')
plt.ylabel('Total Sales Amount')
plt.show()

```
